### Punto 1.

Una pequeña línea de producción consiste en 3 operaciones, cada una con una máquina que procesa una pieza a la vez. En cada operación, la pieza tiene una probabilidad de ser procesada exitósamente, de lo contrario tendrá que ser reprocesada en la misma máquina, según la siguiente tabla de probabilidades:

|        Operación      | Probabilidad de éxito |
| :--------------------: | :-------------:     |
|    1        |       89%           |
| 2                |       75%           |
|3                |       95%           |

Suponga que cada operación demora una unidad de tiempo, sea que la pieza sea procesada por primera vez o sea reprocesada.	
* Cree una función con el nombre de `unaPieza` para simular el recorrido de una pieza desde el comienzo del proceso de producción hasta estar totalmente terminada.
* Utilice la función creada anteriormente para simular el proceso de producción de `n` piezas en la función `Planta` e imprima el tiempo promedio requerido para terminar una pieza con base en los `n` datos.

**Ayuda: El comando `np.random.random()` le ayudará a generar números aleatorios distribuidos uniformemente entre 0 y 1**

**Resultado Esperado**: 
```python
Planta(100) = El tiempo promedio de procesamiento de 100 piezas es 3.5 segundos
```

````python
def unaPieza(pieza,operacion):
        v_tiempo       = 1                       # Unidad de tiempo
        v_prob_exito   = [89,75,95]              # Porcentajes probabilidad de exito definidos por operacion
        v_tiempo_total = 0                       # Tiempo total de la pieza segun su operacion
        v_prob_defini  = v_prob_exito[operacion] # Porcentaje Probabilidad de Exito definido para pieza actual
        v_op_exito     = 'N'                     # Operacion exitosa
       
        while v_op_exito == 'N':
            v_porc_alcanzado = (np.random.rand() * 100)   # Porcentaje calculado para operacion de pieza actual
            v_tiempo_total  += v_tiempo                   # Tiempo total para la pieza, aumenta si debe reprocesarla
            
            if v_prob_defini >= v_porc_alcanzado:
                v_op_exito = 'S'
                
            #print('Pieza {}, Operación {}. Tiempo {} segundo(s). Prob Exito Operacion {}%. Se proceso Exitosamente {}'.format(pieza, operacion + 1, v_tiempo_total, v_prob_defini, v_op_exito))
                
        return v_tiempo_total
````

````python
def planta():
    v_piezas       = np.random.randint(100)  # Cantidad de piezas a procesar aleatoreamente
    v_operacion    = v_tiempo_final = 0
    
    #print('-------- Resultados Linea de Operación --------')
    
    # Recorre ciclo para el numero de piezas a procesar
    for pieza in range(1,(v_piezas + 1)):
        v_tiempo_final += unaPieza(pieza,v_operacion) # Llama función para calculo de tiempo de la pieza
        v_operacion += 1                              # Una vez camba de pieza sigue con siguiente operación
        if v_operacion == 2:                          # Cuando llega a la tercera operacion vuleve a la inicial
            v_operacion = 0            
    
    # Imprime resultados
    #print('-------- Totales --------')
    print('Planta({}) = El tiempo promedio de procesamiento de {} piezas es {} segundo(s)'.format(v_piezas,v_piezas,v_tiempo_final))
````

Escriba una clase llamada `EvenNumbers`. Cada instancia de la clase debe inicializase con 2 parámetros: un número que actuará como límite inferior y uno como límite superior, ambos enteros entre 1000 y 9999. La clase debe contener una función llamada `EvenBetween` que imprima todos los números dentro del rango (límite inferior, límite superior) que cumplan la condición que todos sus digitos sean pares. El resultado de la función debe ser cada uno de los números que cumplan la condición separados por una coma `(,)`

**Resultado Esperado**
```python
mi_clase = EvenNumbers(2000, 2050)
mi_clase.EvenBetween() = '2000, 2002, 2004, 2006, 2008, 2020, 2022, 2024, 2026, 2028, 2040, 2042, 2044, 2046, 2048'
```

````python
class EvenNumbers:
   
    def __init__(self, inferior, superior):
        self.inferior = inferior
        self.superior = superior

    def EvenBetween(self):
        v_numeros = []  # Lista de numeros a imprimir
        for numero in range(self.inferior,self.superior):
            v_impares = 0
            v_numero  = str(numero)
            for digito in v_numero:
                if int(digito) % 2 != 0: # Si el digito es impar suma al conteo de digitos de ese numero
                    v_impares += 1
            
            if v_impares == 0:   # Si no hay impares agrega el número a la lista
                v_numeros.append(str(numero))
                
        print(','.join(v_numeros))  # Imprime lista de numeros
````

La estrategia del equipo es hacer que el robot realice una caminata aleatoria, es decir que con igual probabilidad el robot se desplace a una celda hasta que llegue al objetivo. Ya que el resultado depende la suerte que tuvo el robot realice `n` réplicas para cada uno de los terrenos y reporte el valor esperado de la distancia recorrida.

Su solución debe ser una función o programa que imprima en pantalla el promedio de la distancia recorrida en las `n` replicas de un terreno particular (El terreno y el número de replicas deben entrar por parametro).

**Ayuda: Para leer un archivo `.txt` se puede usar el comando `pd.read_csv(nombre_archivo, header = None, sep = '\t')`**

**Resultado Esperado:**


````python
def Robot(archivo, replicas):
    
    # Lee archivo
    df = pd.read_csv(archivo, header = None, sep = '\t')
    
    numero_filas                = len(df) - 1
    numero_columnas             = len(df.columns) - 1
    x = y = pasos = total_pasos = 0
    posicion_robot              = df[0][0]  
    total_replicas              = 1
        
    while total_replicas <= replicas:
        
        while str(posicion_robot) != 'f': 

            # Genera movimiento aleatorio
            v_mov_aleatorio = np.random.randint(4)
            se_movio        = 0

            # Movimiento derecha
            if v_mov_aleatorio == 0 and x < numero_columnas and str(df[x + 1][y]) != '1':
                x += 1
                se_movio += 1
            # Movimiento izquierda
            elif v_mov_aleatorio == 1 and x > 0 and str(df[x - 1][y]) != '1':
                x -= 1  
                se_movio += 1
            # Movimiento abajo
            elif v_mov_aleatorio == 2 and y < numero_filas and str(df[x][y + 1]) != '1':
                y += 1
                se_movio += 1
            # Movimiento arriba
            elif v_mov_aleatorio == 3 and y > 0 and str(df[x][y - 1]) != '1':
                y -= 1
                se_movio += 1

            # Cada que hace un movimiento
            if se_movio > 0:
                posicion_robot = df[x][y]
                pasos += 1
    
        # Estadisticas de todas las replicas
        total_pasos    = total_pasos + pasos
        promedio_pasos = total_pasos / total_replicas
        total_replicas += 1
    
    #Impresion final
    print('Para el terreno {}, con una muestra de {} replica(s) el robot llega a su destino en {} pasos'.format(archivo[5:6], replicas, promedio_pasos))
````

````python
def ExploracionPetrolera(numero_exploraciones):

    v_diccionario = {"Punto":[],"X":[],"Y":[],"Tamaño":[],"Factibilidad":[] };

    for a in range(numero_exploraciones):
        
        # Calculo de coordenadas
        x = int(np.random.randint(1, 31))
        y = int(np.random.randint(1, 21))
        
        # Validaciones Factibilidad
        if (y <= (2 * np.sin(x - 3) + 17)) and (y >= ((x/3)-5)) and (((x**2) + (y**2)) <= (32**2)):
            factible='Factible'
        else:
            factible='Infactible'
        
        # Calculo tamaño del pozo
        tam_pozo = (((2 * x**2)-(3 * y**2) - np.cos(8 * x * y) + 10 * np.sin(x) * np.cos (x/3))**2)
        
        # Agregar valores a Diccionario
        v_diccionario["Punto"].append(a + 1)
        v_diccionario["X"].append(x)
        v_diccionario["Y"].append(y)
        v_diccionario["Tamaño"].append(int(tam_pozo))
        v_diccionario["Factibilidad"].append(factible)
    
    # Dataframe - ordenar
    df  = pd.DataFrame.from_dict(v_diccionario)
    df1 = df.sort_values(['Factibilidad','Tamaño'], ascending=[True, False])
    
    # Imprime Dataframe
    print(df)
    
    # Imprime pozo mayor tamaño Factible
    print('\nEl pozo a explorar se encuentra en las coordenadas ({}, {}), cuenta con un tamaño de {} y es {}'.format(df1.iat[0, 1], df1.iat[0, 2], df1.iat[0, 3], df1.iat[0, 4]))
````

### Punto 5.

Dos personas están apostando en un juego que consiste en recolectar la mayor cantidad de monedas de 100 pesos. En total se encuentran 20 monedas en la mesa y cada jugador lanza un dado de 6 caras para determinar la cantidad de monedas que puede retirar de la mesa. Ambos jugadores se alternan para lanzar el dado empezando por el  primer jugador. En el caso de que el número del dado sea superior a la cantidad de monedas no se podrá tomar ninguna moneda y se cede el turno al otro jugador.

Implemente una función llamada `Juego` para que imprima en pantalla el jugador que ganó (jugador 1 o jugador 2), la cantidad de monedas obtenidas y el número de turnos que tuvo el juego

**Ayuda: el método `np.random.randint()` le ayudará a generar números aleatorios enteros entre 2 valores**

**Resultado Esperado**: 
```python
Juego() = Despues de 8 turnos, el ganador es el Jugador 2 con un total de 11 monedas
```

````python
def Juego():
    jugador_1 = jugador_2 = turnos_juego = monedas_gana = 0
    monedas = 20
    turno_jugador = 1
   
    # Mientras hayan monedas en la mesa
    while monedas > 0:
        
        turnos_juego += 1
        
        # Numero aleatorio al lanzar el dado
        dado_resultado = np.random.randint(1, 7)
        
        # Si hay mas monedas de lo que se sacó en el dado
        if monedas >= dado_resultado:
            if turno_jugador == 1:
                jugador_1 += dado_resultado
            else:
                jugador_2 += dado_resultado
            
            # Se restan monedas obtenidas
            monedas -= dado_resultado
        
        # Se asigna siguiente turno al siguiente jugador
        turno_jugador += 1
        
        # Si se paso de jugador 2 vuelve al 1
        if turno_jugador > 2:
            turno_jugador = 1
    
    # Revision del ganador
    if jugador_1 > jugador_2:
        gana         = 'el ganador es el Jugador 1'
        monedas_gana = jugador_1
    elif jugador_1 < jugador_2:
        gana         = 'el ganador es el Jugador 2'
        monedas_gana = jugador_2
    elif jugador_1 == jugador_2:
        gana = 'hay empate entre el Jugador 1 y el Jugador 2'
        monedas_gana = jugador_2
    
    # Impresion de resultados
    print('Juego() = Despues de {} turnos, {} con un total de {} monedas'.format(turnos_juego, gana, monedas_gana))
````
